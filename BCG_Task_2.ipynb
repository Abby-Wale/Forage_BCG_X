## sub-task 2
hypothesis verification

```python

# Define a metric for price sensitivity (percentage change in price)
price_df['price_sensitivity'] = (price_df['price_peak_var'] - price_df['price_off_peak_var']) / price_df['price_off_peak_var']

# Merge Dataframes

merged_df = pd.merge(client_df, price_df, on='id')
```

```python
# Calculate Price Sensitivity
merged_df['price_sensitivity'] = (merged_df['price_peak_var'] - merged_df['price_off_peak_var']) / merged_df['price_off_peak_var']

# Correlation Analysis
correlation = merged_df[['price_sensitivity', 'churn']].corr()

# Print the correlation matrix
print("\nCorrelation Matrix:")
correlation
```

```python


# Visualize the correlation using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix - Price Sensitivity and Churn')
plt.show()
```
